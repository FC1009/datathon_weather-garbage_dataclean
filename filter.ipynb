```python
wea_name = "新竹市"
```

```python
df1 = pd.read_csv("./data_clean/data/{}.csv".format(wea_name),sep=",")
# df1 = pd.read_csv("./final_project/test.csv",sep=",",encoding="utf-8")
```

```python
# filter = pd.to_datetime(df1['時間']) >= "2020-01-01-01"
filter = df1['時間'].apply(str) >= "2020-01-01-01"
```

```python
df1 = df1[filter]
```

```python
df1
```

```python
df1 = df1.rename(index=str, columns={"降水量(mm)": "降雨量"})
# filter = pd.to_numeric(df1['降雨量'])
```

```python
df2 = df1[["時間","縣市","觀測站名","降雨量"]]
df2
```

```python
df2 = df2[df2["降雨量"] != "T"]
df2 = df2[df2["降雨量"] != "X"]
# df2 = df2[df2["降雨量"] != "..."]
# df2 = df2[df2["降雨量"] != "\\"]
# df2 = df2[df2["降雨量"] != "/"]
```

```python
df3= df2.groupby(["時間"])["降雨量"].max()
# df2["max_1H"] = df2.groupby(["時間"])["降雨量"].cummax()
# df3 = df2.groupby(["時間"])["降雨量"].rolling(3).sum()
```

```python
df3
```

```python
df4 = pd.DataFrame({'時間':df3.index, '降雨量':df3.values})
df4
```

```python
type(df4['降雨量'][0])
```

```python
df4["Alert"] = 0
df4
```

```python
for i in range(len(df4)):
    if float(df4['降雨量'][i]) >= 1:
        df4["Alert"][i] = 1
df4
```

```python
df4[["年","月","日","小時"]] = df4["時間"].str.split('-',expand=True)
df4["日期"] = df4["年"]+df4["月"]+df4["日"]
df4 = df4.drop(["年","月","日"],axis = 1)
df4
```

```python
df5= df4.groupby(["日期"])["Alert"].rolling(3).sum()
```

```python
df6 = df5.to_frame()
```
